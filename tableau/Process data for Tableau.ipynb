# Process data for Tableau

### Importing required libraries

```python
import pandas as pd
import numpy as np
import datetime
```

### loading csv file

```python
df = pd.read_csv('processed_data_sql.csv')
```

### filtering out the required data

```python
df = df.drop(['Unnamed: 0', 'start_lat','start_lng', 'end_lat', 'end_lng'], axis=1)
```

```python
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])
df['day'] = df['started_at'].dt.day_name()
df['month'] = pd.DatetimeIndex(df['started_at']).month
df['year'] = pd.DatetimeIndex(df['started_at']).year
df['hour of the day'] = df['started_at'].dt.hour
df['day of week'] = df['started_at'].dt.dayofweek
df['duration'] = (df['ended_at'] - df['started_at']).dt.total_seconds() / 60
```

```python
df = df[df.duration > 1]
```

```python
df.dropna(inplace = True)
```

```python
df.drop_duplicates(subset=['ride_id'])
```

### exporting the new data for tableau

```python
df.to_csv('processed_data_tableau.csv', index=False)
```
